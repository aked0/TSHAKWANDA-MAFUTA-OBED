TABLEAU A UNE DIMENSION 

```python
import numpy as np

tab_1D=np.array ([5, 2, 3, 25, 64])
print (f"le tableau à une dimension : {tab_1D}")
```

TABLEAU A DEUX DIMENSIONS

```python
import numpy as np

tab_2D=np.array ([
    [5, 2, 3, 25, 64],
    [58, 3, 4, 17, 33]
])
print (f"le tableau à deux dimensions : {tab_2D}")
```

TABLEAU A 3 DIMENSIONS

```python
import numpy as np

tab_3D=np.array ([[
    [5, 2, 3, 25, 64],
    [58, 3, 4, 17, 33],
    [99, 66, 48, 58, 67]
]])
print (f"le tableau à trois dimensions : {tab_3D}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,2,50)
fig, ax= plt.subplots(figsize=(6,4))
ax.plot(x,x*2, label='double')
ax.plot(x,x**2, label='carré de X')
ax.plot(x,x**3, label='cube de X')
ax.set_xlabel('Données de x')
ax.set_ylabel('Données de y')
plt.title('Données de linspace entre 0 et 2')
ax.legend()
```
